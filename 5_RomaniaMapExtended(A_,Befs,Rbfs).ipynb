Best first search

```python
import heapq
def greedy_best_first_search(graph, initial, goal, heuristic):
    priority_queue = []
    heapq.heappush(priority_queue, (heuristic[initial], [initial]))
    visited = set()

    while priority_queue:
        _, path = heapq.heappop(priority_queue)
        node = path[-1]

        if node in visited:
            continue

        visited.add(node)

        if node == goal:
            return path

        for neighbor in sorted(graph.get(node, {}), key=lambda x: heuristic[x]):
            if neighbor not in visited:
                new_path = path + [neighbor]
                heapq.heappush(priority_queue, (heuristic[neighbor], new_path))

    return None

start = input("Enter starting node: ").capitalize()
des = input("Enter destination node: ").capitalize()

path = greedy_best_first_search(Romania_Map, start, des, heuristic)

if path:
    print("\nPath:")
    total_cost = 0
    for i, node in enumerate(path):
        if i > 0:
            step_cost = Romania_Map[path[i - 1]][node]
            total_cost += step_cost
            print(f"Step {i}: {node} (Step Cost: {step_cost})")
        else:
            print(f"Step {i}: {node}")

    print(f"Total Cost: {total_cost}")
else:
    print(f"No path found from {start} to {des}.")
```

A* search

```python
def astar_search(graph, start, goal, heuristic):
    open_set = [(0 + heuristic[start], start, [])]
    closed_set = set()

    while open_set:
        open_set.sort(key=lambda x: x[0])
        current_f, current_node, current_path = open_set.pop(0)

        if current_node == goal:
            return current_path + [current_node]

        if current_node in closed_set:
            continue

        closed_set.add(current_node)

        for neighbor, cost in graph.get(current_node, {}).items():
            if neighbor not in closed_set:
                g_cost = current_f - heuristic[current_node]
                new_g_cost = g_cost + cost
                new_path = current_path + [current_node]
                open_set.append((new_g_cost + heuristic[neighbor], neighbor, new_path))


start = input("Enter starting node: ").capitalize()
goal = input("Enter destination node: ").capitalize()

path = astar_search(Romania_Map, start, goal, heuristic)

if path:
    print("\nPath:")
    total_cost = 0
    for i, node in enumerate(path):
        if i > 0:
            step_cost = Romania_Map[path[i - 1]][node]
            total_cost += step_cost
            f_cost = total_cost + heuristic[node]
            print(f"Step {i}: {node} (f(n) = g(n) + h(n): {total_cost} + {heuristic[node]} = {f_cost})")
        else:
            print(f"Step {i}: {node} (f(n) = g(n) + h(n): 0 + {heuristic[node]} = {heuristic[node]})")

    print(f"Total Cost: {total_cost}")
else:
    print(f"No path found from {start} to {goal}.")
```

RBFS

```python
def rbfs_search(graph, current, goal, f_limit, heuristic):
    if current == goal:
        return [current], 0

    successors = []
    for neighbor, cost in graph.get(current, {}).items():
        h = heuristic.get(neighbor, 0)
        successors.append((neighbor, cost + h))

    if not successors:
        return [], float('inf')

    while True:
        successors.sort(key=lambda x: x[1])
        best = successors[0]

        if best[1] > f_limit:
            return [], best[1]

        alternative = successors[1][1] if len(successors) > 1 else float('inf')
        result, best_cost = rbfs_search(graph, best[0], goal, min(f_limit, alternative), heuristic)

        if result is not None:
            return [current] + result, best_cost

# Usage
start = input("Enter starting node: ").capitalize()
goal = input("Enter destination node: ").capitalize()

path, total_cost = rbfs_search(Romania_Map, start, goal, float('inf'), heuristic)

if path:
    print("\nPath:")
    for i, node in enumerate(path):
        print(f"Step {i + 1}: {node}")
else:
    print(f"No path found from {start} to {goal}.")
```
